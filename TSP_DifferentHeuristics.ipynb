## Comparing Traveling Salesman Problem : Different Heuristics

```r
#Program in : R 
#Credits  : operatiology
#install.packages("tspmeta")
#before you run this program
```

```r
options(warn=-1)
library(tspmeta)
latlongs = read.csv("latlongs.csv",nrows=50)
coords.df <- data.frame(lon = latlongs$lon, lat = latlongs$lat)
coords.mx <- as.matrix(coords.df)
```

```r
# Compute distance matrix
dist.mx <- dist(coords.mx)
# Construct a TSP object
tsp.ins <- tsp_instance(coords.mx, dist.mx )
```

### Run the solver using 2-opt

```r
tour <- run_solver(tsp.ins, method="2-opt")
```

### Plot the 2-opt tour

```r
autoplot(tsp.ins, tour)
```

### The Heuristics are : 
 - nearest_insertion
 - farthest_insertion
 - cheapest_insertion
 - arbitrary_insertion
 - nearest neighbours 
 - repetitive_nn
 - 2-opt **(demonstrated above)

```r
heus <- c("nearest_insertion", "farthest_insertion",
             "cheapest_insertion", "arbitrary_insertion", "nn", "repetitive_nn",
             "2-opt")
tours <- sapply(heus, FUN = function(m) run_solver(tsp.ins, method = m), simplify = FALSE)
```

### Plot the comparison

```r
dotchart(c(sapply(tours, FUN = attr, "tour_length")),
         xlab = "Tour length", main="Heuristic Solutions")
```
